```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 3.5, 6.67e-6)
freq = 20
dv1 = np.sin(freq * 2 * np.pi * time)
dv1_noise = dv1 + np.random.normal(0, 0.1, len(time))
```

```python
taps = 601
cutoff = [10, 30]
fs = len(time) / time[-1]
```

```python
filters = wtu.FIRBandPass(taps, cutoff, fs)
dv1_filt = filters.filter(dv1_noise)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(time, dv1_noise, 'b', label='Noise')
ax.plot(time, dv1, 'w', label='Original')
ax.plot(time, dv1_filt, '--r', label='filter')
ax.set_xlim(0, 0.5)
ax.legend(loc='upper right')
```
