```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
time = np.arange(0, 3.5, 6.67e-6)
dv1 = np.sin(5 * np.pi * time)
dv1_deriv = 5 * np.pi * np.cos(5 * np.pi * time)
```

```python
dv1_deriv_ana = wtc.derivitive(dv1, time)
```

```python
# Work on normalization
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(time, dv1, 'k')
ax.plot(time, dv1_deriv_ana, 'r')
ax.plot(time, dv1_deriv, '--k')
ax.set_xlim(0, 3.5)
plt.tight_layout()
```
