```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_sections(wave_sec, wave, time, num_plots=5):
    sparse = len(wave_sec) // num_plots + 1
    fig, ax = plt.subplots(num_plots, sharey=True, figsize=(10, 5*num_plots))
    i = 0
    for start, end in wave_sec[::sparse]:
        ax[i].plot(time[start:end], wave[start:end])
        i += 1
```

```python
time = np.arange(0, 3.5, 6.67e-6)
v12_delay = 0.0
v34_delay = 0.0
freq = 700
dv1 = np.sin(freq * 2 * np.pi * time) +  0.4 * np.sin(2 * freq * 2 * np.pi * time)
dv2 = np.sin(freq * 2 * np.pi * (time + v12_delay))
dv3 = np.sin(freq * 2 * np.pi * time)
dv4 = np.sin(freq * 2 * np.pi * (time + v34_delay))
```

```python
single_freq = wtu.frequency_filter(dv1, dv2, dv3, dv4, time)
```

```python
plot_sections(single_freq, dv1, time)
```
