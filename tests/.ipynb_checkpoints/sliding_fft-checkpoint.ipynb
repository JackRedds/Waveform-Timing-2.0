```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
```

```python
time = np.arange(0, 3.5, 6.67e-6)
freq = 1500
dv1 = np.sin(freq * 2 * np.pi * time)
```

```python
spec, time_spec, xf, __ = wtu.sliding_fft(dv1, time, window_size=10000)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

im = ax.pcolormesh(time_spec, xf, spec.T, shading='gouraud', cmap='jet', norm=mpl.colors.LogNorm())
cb = fig.colorbar(im)
cb.set_label('dV1', size='large')
ax.set_ylabel('Frequency [Hz]')
ax.set_xlabel('time [s]')
ax.axhline(freq, c='k')
ax.set_ylim(0, 3000)
plt.tight_layout()
```
