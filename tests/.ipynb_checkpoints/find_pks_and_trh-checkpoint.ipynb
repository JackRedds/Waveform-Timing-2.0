```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 3.5, 6.67e-6)
dv1 = np.sin(5 * np.pi * time)
```

```python
peaks, troughs = wtu.find_peaks_and_troughs(dv1)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(time, dv1, 'k')
ax.plot(time[peaks], dv1[peaks], 'ro')
ax.plot(time[troughs], dv1[troughs], 'ro')
ax.set_xlim(time.min(), time.max())
plt.tight_layout()
```
