```python
import wave_timing.calc as wtc
import wave_timing.utils as wtu
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
time = np.arange(0, 3.5, 6.67e-6)
dv1 = np.sin(20 * np.pi * time)
dv2 = np.sin(20 * np.pi * (time + 0.02))
dv3 = np.sin(20 * np.pi * time)
dv4 = np.sin(20 * np.pi * (time + 0.004))
```

```python
delays = wtc.multiple_delay(dv1, dv2, dv3, dv4, time)
peaks, troughs = delays.peaks_and_troughs(start=time[0], end=time[-1])
```

```python
fig, ax = plt.subplots()
ax.plot(time, dv1)
ax.plot(time[peaks[0]], dv1[peaks[0]], '.')
ax.plot(time[troughs[0]], dv1[troughs[0]], '.')
plt.tight_layout()
```

```python
multi_delays_12, ave_pos_12 = delays.delay(peaks[0], peaks[1])
multi_delays_34, ave_pos_34 = delays.delay(peaks[2], peaks[3])
```

```python
fig, ax = plt.subplots()
ax.plot(ave_pos_12 * delays.dt, multi_delays_12 * delays.dt, '.')
ax.set_ylim(-0.1, 0.1)
plt.tight_layout()
```

```python
print(f'Length of V1-2 Delays Before Trim: {len(multi_delays_12)}')
print(f'Length of V3-4 Delays Before Trim: {len(multi_delays_34)}')

same_len_delays, same_len_times = delays.trim_delays(multi_delays_12, ave_pos_12, multi_delays_34, ave_pos_34)


print(f'Length of V1-2 Delays After Trim: {len(same_len_delays[0])}')
print(f'Length of V3-4 Delays After Trim: {len(same_len_delays[1])}')
```

```python
delay_peaks, delay_pos_peaks, delay_troughs, delay_pos_troughs = delays.calculate_delays()
```

```python
fig, ax = plt.subplots(2, 2, sharex='col', sharey='row', figsize=(20, 10))

ax[0, 0].plot(time, dv1)
ax[0, 0].plot(time, dv2)
ax[1, 0].plot(delay_pos_peaks[0], delay_peaks[0], '.')
ax[1, 0].set_ylabel(f'Ave Delay: {round(np.mean(delay_peaks[0]), 4)}')

ax[0, 1].plot(time, dv3)
ax[0, 1].plot(time, dv4)
ax[1, 1].plot(delay_pos_peaks[1], delay_peaks[1], '.')
ax[1, 1].set_ylabel(f'Ave Delay: {round(np.mean(delay_peaks[1]), 4)}')

ax[1, 0].set_ylim(-0.1, 0.1)

plt.tight_layout()
```
